```python
import nltk
from nltk.corpus import words
import re
from nltk.tokenize import word_tokenize
from collections import defaultdict
```

```python
from pyspark import SparkContext

sc_1 = SparkContext("local", "WordCount")

fil = sc_1.textFile("file1.txt")

c_words = fil.flatMap(lambda line: line.split()) \
                   .map(lambda word: (word.lower(), 1)) \
                   .reduceByKey(lambda a, b: a + b) \
                   .collect()

for word, count in c_words:
    print(f"{word}: {count}")

sc_1.stop()
```

```python
#question 2

nltk.download('punkt')
nltk.download('words')

def read_file(filename):
    with open('file2.txt', 'r') as file:
        return file.read()

def detect_words(text):
    sc_2 = defaultdict(int)
    words_english = set(nltk.corpus.words.words())

    words = word_tokenize(text.lower())

    for word in words:
        word = re.sub(r'[^\w\s]', '', word)
        if word and word not in words_english:
            sc_2[word] += 1

    return sc_2

def map_reduce(filename):
    text = read_file(filename)
    num_counts = detect_words(text)
    return num_counts

if __name__ == "__main__":
    num_counts = map_reduce("Chapter_text_2.txt")

    print("Non-English Words and Their Counts:")
    for word, count in sorted(num_counts.items()):
        print(f"{word}: {count}")
```
